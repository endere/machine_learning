```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv("Building_Permits___Current.csv")
print(df)
```

```python
#I am curious about the difference between the average and median values of the building permits in seattle.
#At first glance, I see a ton of values at 0, so there's a chance those could substantially drag down the average.
#However, as I cannot see all of the permits, I'm guessing there is a few gigantic permits that will drastically raise
#The average far above the median.
df['Value'].map(lambda x: float(x.lstrip('$'))).median()
#In order to get this figure, I had to strip the dollar sign from each figure with a lamda function.
```

```python
df['Value'].map(lambda x: float(x.lstrip('$'))).mean()
#Sure enough, the mean is far greater than the median. So while the vast majority of the permits are lower cost,
#A few more expensive ones are drastically raising the average.
```

```python
#Let's see what that most expensive permit is, dragging up that average...
max(df['Value'].map(lambda x: float(x.lstrip('$'))))
#292 million dollars... That is quite the sum...
```

```python
#Now.. what kind of place is it?
largest = df['Value'].map(lambda x: float(x.lstrip('$'))).idxmax()
print(df['Address'][largest])
print(df['Permit Type'][largest])
print(df['Category'][largest])
print(df['Description'][largest])
#A gigantic high rise building... That is not particularly surprising. I looked it up on google maps, and sure enough,
#There's a huge construction zone there right now visible on google street view.
```

```python
#I wanted to know the standard amount of time between applying for a building permit and actually receiving one.
#There were a ton of NaNs in both of the columns I used, and, as those were impossible to use, I wrote this code to
#automatically discard them. This is the largest amount of data cleaning that I have done in this excersize thus far.
import datetime
dated_apps = df['Application Date'].map(lambda x: datetime.datetime.strptime(x, '%m/%d/%Y') if type(x) == str else datetime.datetime.strptime('1/1/1000', '%m/%d/%Y'))
dated_issues = df['Issue Date'].map(lambda x: datetime.datetime.strptime(x, '%m/%d/%Y') if type(x) == str else datetime.datetime.strptime('1/1/1000', '%m/%d/%Y'))
dates = dated_issues - dated_apps
time_delta_data = pd.Series(list(filter(lambda date: date.days < 50000 and date.days > -1, dates)))
print(time_delta_data)
#Pretty cool numbers! I'll poke around with them in the next cell, in order to keep things well separated.
```

```python
#The easiest first question to ask when examining how long it takes to get a permit is the mean and median.
print(time_delta_data.median())
print(time_delta_data.mean())
#Seems like either most cases have a pretty quick turn around time. That, or due to human lazyness and forgetfulness,
#People write in the same date on both columns of their documents, which causes a lot of this data to be useless.
```

```python
#So... median at zero means a lot of these dates are same day issuings. That seems like a lot. What percentage of 
#these columns are the same?
same = list(filter(lambda x: x.days == 0, time_delta_data))
print(len(same) / len(time_delta_data) * 100, '%')
#Just north of half. Either beurocracy works quickly, and people can instantly get permits in Seattle these days,
#or some people have been lazy about their book keeping. Either way, I am far from qualified to comment on it.
```

```python
#Looking more into this... If we filtered out all of those zeros, how does our median and mean look?
time_delta_data_minus_zeros = pd.Series(list(filter(lambda date: date.days < 50000 and date.days > 0, dates)))
print(time_delta_data_minus_zeros.median())
print(time_delta_data_minus_zeros.mean())
#Certainly more realistic numbers, in my limited perspective. However, that would be throwing out data, and is thus
#bad practice, even if the data itself is theoretically not great. This is, unfortunately, unusable data to me.
#That's not to say it is completely unusable, I am just not a trained data scientist, and don't know the first thing
#of the nature of the data that I am examining.
```
