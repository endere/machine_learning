```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv("boston_housing_data.csv", delim_whitespace=True, names=["crim","zn","indus","chas","nox","rm","age","dis","rad","tax","ptratio","black","lstat","medv"])
print(df)
```

```python
df['medv'].median()
```

```python
df['medv'].mean()
```

```python
#These previous two functions mean that, on average, most homes in the area are around 21-22 thousand dollars.
#However, as the median is less than the mean, this means that a majority of the houses are on the lower side of that spectru,
#with a few more expensive houses bringing up the average.
```

```python
df['medv'][df['age'].idxmin()]
#Houses are substantially more expensive (on average) in the area with the newest homes.
#A safe guess says this is because they are nicer, due to how new they are.
```

```python
#crime rate versus pupil teacher ratio. Dots are blue if suburb is bounded by the Charles river.
x = df['crim']
y = df['ptratio']
z = df['chas']
colors = ['blue' if df['chas'][i] == 1 else 'red' for i in range(506)]
plt.scatter(x, y, color=colors)
plt.xlabel("Per capita crime rate")
plt.ylabel("Pupil teacher ratio")
plt.show()
#There seems to be a HEAVY correlation between high crime rates and exactly 21 students to a teacher.
#That is what this data claims. In reality however, this data is likely skewed for a multitude reasons, and it
#is not possible to make accurate inferences about this correlation from this data.
```

```python
#Proportion of blacks per town versus distance from employment centers.
x = df['black']
y = df['dis']
plt.scatter(x, y)
plt.xlabel("Proportion of black people. 1000(Bk - 0.63)^2")
plt.ylabel("Distance to employment centers.")
plt.show()
#This data seems to show a correlation between the black population and proximity to employment centers.
#However, as most of the points of data fall on the extreme end of the black population line, it may just be that
#there is more data to fill all parts of the y axis there. It is hard to create a well informed assumption with this
#data set.
```

```python
x = df['medv']
y = df['nox']
z = df['indus']
plt.subplot(1, 2, 1)
plt.scatter(y, x)
plt.ylabel("Median Home Value")
plt.xlabel("NO concentration (parts per 10M)")
plt.subplot(1, 2, 2)
plt.scatter(z, x)
plt.xlabel("Proportion of Non retail Businesses")
plt.show()
#These data sets seem similarly shaped, and there may be some correlation represented here.
#I do not claim to understand how nitrogen oxides work, and what they may have to do with non retail businesses or home values.
```

```python
#I have no hypotheses to confirm here. I am just curious about how buildings may have changed over time with regards
#to how many rooms they have on average since the 1940s.
x = df['age']
y = df['rm']
plt.scatter(x, y)
plt.xlabel("Proportion of buildings built prior to 1940.")
plt.ylabel("Average number of rooms per dwelling.")
plt.show()
#This data shows that more buildings with fewer rooms were built prior to 1940. However, as the majority of the data
#weighs towards the older end of the spectrum, there is a strong chance that it may just be more data represented for
#all sizes of rooms, and thus not significant in actually giving useful information about the question.
```
